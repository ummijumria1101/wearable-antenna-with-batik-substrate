### PENGUKURAN 1A (2,688 GHz)


*   fr: 2.688
*   s21: -2.869
*   tebal/d: 0.23
*   er: 2.083





DATA PENGUKURAN

```python
fr = 2.4*(10**9) #Hz
fr_loss = 2.688 #GHz
s21 = -2.869
d = 0.23 *(10**(-3)) #m
er = 2.083
c = 3*(10**8) # m/s
zo = 50 # ohm
```

PATCH

```python
# Lebar Patch
x1 = 2/(er+1)
W = (c/(2*fr))*(np.sqrt(2/(er+1)))

# Eeff - Leff - Delta L
E_eff = ((er+1)/2)+((er-1)/2)*((1+(12*d/W))**-0.5)
L_eff = c/(2*fr*np.sqrt(E_eff))
x2 = (E_eff+0.3)*((W/d)+0.264)
x3 = (E_eff-0.258)*((W/d)+0.8)
Del_L = 0.412*d*(x2/x3)

# Panjang Patch
L = L_eff-(2*Del_L)
```

```python
W
```

```python
E_eff
```

```python
L_eff
```

```python
Del_L
```

```python
L
```

GROUNDPLANE

```python
# Panjang Groundplane
Lg = (6*d)+L

# Lebar Groundplane
Wg = (6*d)+W
```

```python
Lg
```

```python
Wg
```

FEEDLINE

```python
# B - LamdaG
B = (60*(np.pi**2))/(zo*np.sqrt(er))
Lamda_O = c/fr
Lamda_G = Lamda_O/(np.sqrt(E_eff))

# Lebar Feed
Wf = (2*d/np.pi)*((B-1-np.log((2*B)-1))+(((er-1)/(2*er))*(np.log(B-1)+0.39-(0.61/er))))

# Panjang Feed
Lf = Lamda_G/4
```

```python
Lf
```

```python
Wf
```

INSERT FEED

```python
# Panjang Insert Feed
Lif = Lf/2

# Lebar Insert Feed
Wif = Wf/4
```

```python
Lif
```

```python
Wif
```

LOSS TANGENT

```python
# loss tangent
var_1 = s21/20
var_2 = 10**var_1

r_total_loss = (2*zo*(1-var_2))/(var_2)
loss_tangent = (565 - r_total_loss - 455*(2.71-fr_loss))/(3591.92)
```

```python
r_total_loss
```

```python
loss_tangent
```
