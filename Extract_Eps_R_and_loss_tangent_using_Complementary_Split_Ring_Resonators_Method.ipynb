```python
#eps r
import math
Spar = -5.055 #dalam db
freq = 1.902 #Ghz
dmut = 1.6

varA = 0.019-0.0175*math.exp(-dmut/0.67)
er = (((freq**-2)-0.137)/varA) +2
er
```

image.png

```python
#loss tangent
s21 = -5.055 #dalam db
z0 = 50 #ohm
varA = s21/20
varB = 10**varA

r_total_loss = (2*z0*(1-varB))/(varB)

loss_tangent = (565 - r_total_loss - 455*(2.71-freq))/(3591.92)

print("R total loss :", r_total_loss)
print('Loss tangent :', loss_tangent)
```
