```python
#cari n

m=0
lmda = 300000000/(frekuensi*(10**9)) #meter
k0 = 2*np.pi/(lmda) #konstanta fasanya
eink_len=len(lnimag)
d_mikrostrip = 1/1000 #m
n_re=[0]*eink_len
n_im=[0]*eink_len
def closest(lst, K):

    return lst[min(range(len(lst)), key = lambda i: abs(K-lst[i]))]

for x in range(eink_len):
  nm0=(1/(k0[x]*d_mikrostrip))*((lnimag[x]+2*0*3.14))
  nm1=(1/(k0[x]*d_mikrostrip))*((lnimag[x]+2*1*3.14))
  nm2=(1/(k0[x]*d_mikrostrip))*((lnimag[x]+2*2*3.14))
  nm3=(1/(k0[x]*d_mikrostrip))*((lnimag[x]+2*3*3.14))
  n_im[x]=(1/(k0[x]*d_mikrostrip))*(-1*lnreal[x])
  if x == 0:
    n0=(1/(k0[0]*d_mikrostrip))*((lnimag[0]+2*m*3.14))
    n_re[x]=n0
  else:

    lst = [nm0,nm1,nm2,nm3]
    K = n_re[x-1]
    n_re[x]=closest(lst, K)
print(n_re)
print(n_im)


```

```python
def Single_Rectangular_Patch(c, fr, er, h):
  W = c/(2*fr*np.sqrt((er+1)/2)) # Width of the patch
  Eeff = ((er+1)/2)+(((er-1)/2)*((1/(np.sqrt(1+(12*(h/W))))))) # Effective Dielectric Constant
  L_eff = c/(2*fr*np.sqrt(Eeff)) # Effective length
  Teta_L = 0.412*h*(((Eeff+0.3)*((W/h)+0.264))/((Eeff-0.258)*((W/h)+0.8))) # length extension (Î”L)
  L = L_eff-(2*Teta_L) # length of the patch
  Lambda_G = c/(fr*(np.sqrt(Eeff)))
  #zo = (120*np.pi)/(np.sqrt(Eeff)*((W/h)+1.393+(0.667*np.log((W/h)+1.44))))
  return W, Eeff, L_eff, Teta_L, L, Lambda_G
```

```python
def Groundplane(h, L, W):
  Lg = 6*h + L
  Wg = 6*h + W
  return Lg, Wg
```

```python
def Microstrip_Feedline(zo, er, h, L, Lambda_G):
  B = (60*(np.pi**2))/(zo*np.sqrt(er))
  Wf = (2*h/np.pi)*((B-1-np.log((2*B)-1))+(((er-1)/(2*er))*(np.log(B-1)+0.39-(0.61/er))))
  Eeff_feed = ((er+1)/2)+(((er-1)/2)*((1/(np.sqrt(1+(12*(h/Wf)))))))
  Lambda_G_feed = c/(fr*(np.sqrt(Eeff_feed)))
  Lf = Lambda_G_feed/4
  return B, Wf, Eeff_feed, Lambda_G_feed, Lf
```

```python
fr = frdata # Hz
er = epsredatabatik/100
c = 3*(10**8) # m/s
h = 0.5*(10**(-3)) # m
zo = 50 #Ohm

print("================================================")
print("Diketahui: ")
print("Frekuensi Tengah\t\t: {:.2f} GHz".format(fr))
print("Kecepatan Cahaya\t\t: {:d} m/s".format(c))
print("Impedansi Input\t\t\t: {:d} Î©".format(zo))
print("Konstanta Dielektrik bahan batik: {:.10f}".format(er))
print("Tinggi Dielektrik bahan batik\t: {:.1f} mm".format(h*(10**3)))


print("================================================")
print("           Single Rectangular Patch")
print("")
W, Eeff, L_eff, Teta_L, L, Lambda_G= Single_Rectangular_Patch(c, fr, er, h)
print("Lebar Patch\t\t\t: {:.5f} mm".format(W*1000))
print("Eeff\t\t\t\t: {:.5f}".format(Eeff))
print("L_eff\t\t\t\t: {:.5f} mm".format(L_eff*1000))
print("Teta_L\t\t\t\t: {:.5f} mm".format(Teta_L*1000))
print("Panjang Patch\t\t\t: {:.5f} mm".format(L*1000))
print("Lambda_G\t\t\t: {:.5f} mm".format(Lambda_G*1000))

print("================================================")
print("                 Groundplane")
print("")
Lg, Wg = Groundplane(h, L, W)
print("Lebar Groundplane\t\t: {:.5f} mm".format(Wg*1000))
print("Panjang Groundplane\t\t: {:.5f} mm".format(Lg*1000))

print("================================================")
print("              Microstrip Feedline")
print("")
B, Wf, Eeff_feed, Lambda_G_feed, Lf = Microstrip_Feedline(zo, er, h, L, Lambda_G)
print("B\t\t\t\t: {:.5f}".format(B))
print('Lebar Microstrip Feedline\t: {:.5f} mm'.format(Wf*1000))
print('Eeff_feed\t\t\t: {:.5f}'.format(Eeff_feed))
print('Lambda_G_feed\t\t\t: {:.5f} mm'.format(Lambda_G_feed*1000))
print('Panjang Microstrip Feedline\t: {:.5f} mm'.format(Lf*1000))

print("================================================")
```
